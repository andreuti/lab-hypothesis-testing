**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
# Filtrar datos
hp_dragon = df[df["Type 1"] == "Dragon"]["HP"].dropna()
hp_grass = df[df["Type 1"] == "Grass"]["HP"].dropna()

# Prueba de hip√≥tesis: media Dragon > media Grass
t_stat, p_value = st.ttest_ind(hp_dragon, hp_grass, alternative='greater')

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Rechazamos H‚ÇÄ: Los Dragon tienen m√°s HP que los Grass (significativo al 5%)")
else:
    print("No se puede rechazar H‚ÇÄ: No hay evidencia suficiente de que los Dragon tengan m√°s HP que los Grass")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# Separar grupos
legendary = df[df["Legendary"] == True]
non_legendary = df[df["Legendary"] == False]

# Stats a comparar
stats = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]

# Probar cada atributo
for stat in stats:
    stat_legend = legendary[stat].dropna()
    stat_nonlegend = non_legendary[stat].dropna()

    t_stat, p_value = st.ttest_ind(stat_legend, stat_nonlegend, alternative='two-sided')

    print(f"\nüîç {stat}")
    print(f"  T-statistic: {t_stat:.4f}")
    print(f"  P-value: {p_value:.4f}")

    if p_value < 0.05:
        print("  ‚û§ Diferencia significativa entre Legendarios y No Legendarios")
    else:
        print("  ‚û§ No hay diferencia significativa")
```

**Challenge 2**

In this challenge, we will be working with california-housing data. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

```python
# Coordenadas
school_coords = (-118, 34)
hospital_coords = (-122, 37)
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
# Funci√≥n de distancia eucl√≠dea

def euclidean_distance(lat1, lon1, lat2, lon2):
    return np.sqrt((lat1 - lat2)**2 + (lon1 - lon2)**2)
```

```python
# Calculo distancia m√≠nima a una escuela o hospital

df["distance_to_school"] = euclidean_distance(df["latitude"], df["longitude"], school_coords[1], school_coords[0])
df["distance_to_hospital"] = euclidean_distance(df["latitude"], df["longitude"], hospital_coords[1], hospital_coords[0])

# Usar la distancia m√≠nima a cualquiera de los dos
df["min_distance"] = df[["distance_to_school", "distance_to_hospital"]].min(axis=1)
```

```python
# Clasificar viviendas cerca (<0.5) vs lejos (>= 0.5)

df["is_close"] = df["min_distance"] < 0.5
```

```python
from scipy.stats import ttest_ind
```

```python
# Prueba de hip√≥tesis: ¬øcasas cercanas son m√°s caras?

close_prices = df[df["is_close"]]["median_house_value"]
far_prices = df[~df["is_close"]]["median_house_value"]

t_stat, p_value = ttest_ind(close_prices, far_prices, alternative='greater')

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Rechazamos H‚ÇÄ: Las casas cercanas son significativamente m√°s caras.")
else:
    print("No se puede rechazar H‚ÇÄ: No hay evidencia suficiente de que las casas cercanas sean m√°s caras.")
```
